```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime,timedelta
import time
from collections import Counter
# First Deals with Missing Data
dataset = pd.read_csv('train.csv')
dataset1=pd.read_csv('test.csv')
dataset=dataset[dataset["purchase_date"].notnull()]
def kesson_table(df):
        null_val = df.isnull().sum()
        percent = 100 * df.isnull().sum()/len(df)
        kesson_table = pd.concat([null_val, percent], axis =1)
        kesson_table_ren_columns = kesson_table.rename(
        columns = {0 : 'Missing Data', 1 :'%'})
        return kesson_table_ren_columns
```

## Delete the outlier

```python
def detect_outliers(df,n, features):
    outlier_indices = []
    for col in features:
        Q1 = np.percentile(df[col], 25)
        Q3 = np.percentile(df[col], 75)
        IQR = Q3 - Q1
        outlier_step = 1.5 * IQR
        outlier_list_col = df[(df[col] < Q1 - outlier_step) | (df[col] > Q3 + outlier_step)].index
        outlier_indices.extend(outlier_list_col)
    outlier_indices = Counter(outlier_indices)
    multiple_outliers = list( k for k, v in outlier_indices.items() if v > n)
    return multiple_outliers
```

```python
Outliers_to_drop = detect_outliers(dataset, 2, ["total_positive_reviews"])
```

## Deals with is_free: Encoder

```python
def deal_with_is_free(x):
    if x == False:
        return 0
    else:
        return 1
all["is_free"]=all["is_free"].apply(deal_with_is_free)
```

```python
all.reset_index(inplace=True)
```

## Deal with 3 cates

```python
# 对genres进行处理
a=all["genres"].str.get_dummies(sep=",")
a["index"]=all["index"]
all=pd.merge(all,a,how="left",on="index")

# 对categories进行处理
a=all["categories"].str.get_dummies(sep=",")
a["index"]=all["index"]
all=pd.merge(all,a,how="left",on="index")

# 对tags进行处理
a=all["tags"].str.get_dummies(sep=",")
a["index"]=all["index"]
all=pd.merge(all,a,how="left",on="index")

#drop掉以前的东西
all.drop(["index","tags","genres","categories"],1,inplace=True)
```

## year，month，date

```python
all["release_year"]=all["release_date"].dt.year
all["release_month"]=all["release_date"].dt.month
all["release_day"]=all["release_date"].dt.day
all["purchase_year"]=all["purchase_date"].dt.year
all["purchase_month"]=all["purchase_date"].dt.month
all["purchase_day"]=all["purchase_date"].dt.day
all["day_interval"]=all["purchase_date"]-all["release_date"]
all["year_interval"]=all["purchase_year"]-all["release_year"]
#convert them to integer
all["day_interval"]=(all["day_interval"] / np.timedelta64(1, 'D')).astype(int)
```

```python
all["day_interval"]=all["day_interval"].abs()
all["year_interval"]=all["year_interval"].abs()
```

```python
all["day_interval"]=all["day_interval"]+1
all["year_interval"]=all["year_interval"]+1
```

## Dealing with reviews

```python
all["rate_total"]=all["total_positive_reviews"]+all["total_negative_reviews"]
all["good_rate"]=all["total_positive_reviews"]/all["rate_total"]
all["bad_rate"]=all["total_negative_reviews"]/all["rate_total"]
all.loc[all["rate_total"]==0,"good_rate"]=0
all.loc[all["rate_total"]==0,"bad_rate"]=0
all["positive_reviews_per_day"]=all["total_positive_reviews"]/all["day_interval"]
all["negative_reviews_per_day"]=all["total_negative_reviews"]/all["day_interval"]
all["positive_reviews_per_year"]=all["total_positive_reviews"]/all["year_interval"]
all["negative_reviews_per_year"]=all["total_negative_reviews"]/all["year_interval"]
```
